```python
class IterativeSegmentation:
    def __init__(
        self, 
        data: np.ndarray, 
        segmentation_method: Callable = KMeans,
        scoring_method: Callable = silhouette_score,
        max_clusters: int = 10,
        min_cluster_separation: float = 0.5
    ):
        """
        Initialize the Iterative Segmentation class.
        
        Parameters:
        -----------
        data : np.ndarray Input data to be segmented
        segmentation_method : Callable, optional Clustering algorithm to use (default is KMeans)
        scoring_method : Callable, optional Metric to evaluate cluster separation (default is silhouette_score)
        max_clusters : int, optional Maximum number of clusters to form (default is 10)
        min_cluster_separation : float, optional Minimum acceptable cluster separation score (default is 0.5)
        """
        super().__init__()
        self.original_data = data
        self.segmentation_method = segmentation_method
        self.scoring_method = scoring_method
        self.max_clusters = max_clusters
        self.min_cluster_separation = min_cluster_separation
        
        # Store results
        self.final_clusters = None
        self.cluster_details = []

    def evaluate_clustering(self, labels: np.ndarray, data: np.ndarray) -> float:
        """
        Evaluate the quality of clustering using the specified scoring method.        
        Parameters:
        -----------
        labels : np.ndarray             Cluster labels
        data   : np.ndarray             Data points        
        Returns:
        --------
        float                           Clustering score
        """
        try:
            return self.scoring_method(data, labels)
        except Exception as e:
            print(f"Scoring error: {e}")
            return 0
        
    def parwise_scoring(self, data: np.ndarray, lab_col : str) -> float:
        """
        calculate the pairwise silhoitte score"""
        labels = sorted(data['labels'].unique())
        pairs = list(itertools.combinations(labels, 2))
        scores = {}
        for i in pairs:
            pair_seg_data = seg_data[seg_data[lab_col].isin(list(i))].reset_index()
            pair_labels = pair_seg_data[lab_col]
            pair_seg_data = pair_seg_data.drop([lab_col], axis = 1)
            scores[i] = self.evaluate_clustering(pair_labels.values, pair_seg_data.values)
            print(scores)
        
        scores = dict(sorted(scores.items(), key=lambda item: item[1]))
        return scores

    def recursive_segment(self, data: np.ndarray, current_depth: int = 0, max_depth: int = 4,
                          threshold = 0.3) -> List[np.ndarray]:
        """
        Recursively segment data, breaking down clusters that are not well-separated.        
        Parameters:
        -----------
        data : np.ndarray                        Data to segment
        current_depth : int, optional            Current recursion depth
        max_depth : int, optional                Maximum recursion depth        
        Returns:
        --------
        List[np.ndarray]                         List of cluster labels
        """
        org_labels = data['labels'].unique()

        if current_depth >= max_depth or data.shape[0] <= 2:
            return [np.zeros(data.shape[0], dtype=int)]
        
        print('Calculating Pairwise score...')
        pairwise_score = self.parwise_scoring(data, 'labels')

        # filter scores for given threshold
        filter_scores = filter(lambda x: x[1] <= threshold, pairwise_score.items())
        if filter_scores == {}:
            print('All Clusters are well seperated according to given threshold.. try different threshold or choose base clustring')
            return data['labels']
        
        # Initialize the list to store labels
        final_labels = np.zeros(data.shape[0], dtype=int)
        
        # while not any([data.shape[0] == 1,]):
        for clust_pair, score in filter_scores.items():
            if len(data['labels'].unique()) >= self.max_clusters:
                print('Clusters are exceeding maximum clusters limit.. making exit')
                return org_labels
            
            clust_algo = self.segmentation_method
            seg_data = data[data['labels'].isin(clust_pair)].reset_index()
            seg_data = seg_data.drop(['labels'], axis = 1)
            print('Performing Clustering...')
            
            for clust_n in range(3, max_depth):
                clust_algo.fit(seg_data, n_clusters = clust_n)
                current_labels = clust_algo.labels_
                seg_data[f'labels_{clust_pair}'] = current_labels
   
                # Check if the current clustering satisfies the threshold
                new_pairwise_score = self.pairwise_scoring(seg_data, labels=f'labels_{clust_pair}')
                if all(score > threshold for score in new_pairwise_score.values()) or \
                                                (org_labels - 2 + clust_n) >= self.max_clusters:
                    break
            
            
            
```

```python



class IterativeSegmentation:
    def __init__(
        self, 
        data: np.ndarray, 
        segmentation_method: Callable = KMeans,
        scoring_method: Callable = silhouette_score,
        max_clusters: int = 10,
        min_cluster_separation: float = 0.5
    ):
        """
        Initialize the Iterative Segmentation class.
        
        Parameters:
        -----------
        data : np.ndarray
            Input data to be segmented
        segmentation_method : Callable, optional
            Clustering algorithm to use (default is KMeans)
        scoring_method : Callable, optional
            Metric to evaluate cluster separation (default is silhouette_score)
        max_clusters : int, optional
            Maximum number of clusters to form (default is 10)
        min_cluster_separation : float, optional
            Minimum acceptable cluster separation score (default is 0.5)
        """
        self.original_data = data
        self.segmentation_method = segmentation_method
        self.scoring_method = scoring_method
        self.max_clusters = max_clusters
        self.min_cluster_separation = min_cluster_separation
        
        # Store results
        self.final_clusters = None
        self.cluster_details = []
    
    def _evaluate_clustering(self, labels: np.ndarray, data: np.ndarray) -> float:
        """
        Evaluate the quality of clustering using the specified scoring method.
        
        Parameters:
        -----------
        labels : np.ndarray
            Cluster labels
        data : np.ndarray
            Data points
        
        Returns:
        --------
        float
            Clustering score
        """
        try:
            return self.scoring_method(data, labels)
        except Exception as e:
            print(f"Scoring error: {e}")
            return 0
    
    def _recursive_segment(
        self, 
        data: np.ndarray, 
        current_depth: int = 0, 
        max_depth: int = 3
    ) -> List[np.ndarray]:
        """
        Recursively segment data, breaking down clusters that are not well-separated.
        
        Parameters:
        -----------
        data : np.ndarray
            Data to segment
        current_depth : int, optional
            Current recursion depth
        max_depth : int, optional
            Maximum recursion depth
        
        Returns:
        --------
        List[np.ndarray]
            List of cluster labels
        """
        if current_depth >= max_depth or data.shape[0] <= 2:
            return [np.zeros(data.shape[0], dtype=int)]
        
        # Try different numbers of clusters
        best_labels = None
        best_score = -1
        best_n_clusters = 1
        
        for n_clusters in range(1, min(self.max_clusters + 1, data.shape[0])):
            # Perform clustering
            clusterer = self.segmentation_method(n_clusters=n_clusters)
            labels = clusterer.fit_predict(data)
            
            # Evaluate clustering
            score = self._evaluate_clustering(labels, data)
            
            # Update best clustering if score is better
            if score > best_score:
                best_score = score
                best_labels = labels
                best_n_clusters = n_clusters
        
        # Check if clustering is satisfactory
        if best_score < self.min_cluster_separation:
            # Recursively segment if clusters are not well-separated
            refined_clusters = []
            for cluster_id in range(best_n_clusters):
                cluster_data = data[best_labels == cluster_id]
                
                # Recursively segment this cluster
                sub_clusters = self._recursive_segment(
                    cluster_data, 
                    current_depth + 1, 
                    max_depth
                )
                
                # Adjust sub-cluster labels
                adjusted_sub_clusters = [
                    sub_cluster + len(refined_clusters) 
                    for sub_cluster in sub_clusters
                ]
                refined_clusters.extend(adjusted_sub_clusters)
            
            return refined_clusters
        
        return [best_labels]
    
    def segment(self) -> List[int]:
        """
        Perform iterative segmentation on the data.
        
        Returns:
        --------
        List[int]
            Cluster labels for each data point
        """
        # Perform recursive segmentation
        cluster_results = self._recursive_segment(self.original_data)
        
        # If multiple cluster configurations are returned, flatten them
        if len(cluster_results) > 1:
            final_labels = np.concatenate(cluster_results)
        else:
            final_labels = cluster_results[0]
        
        self.final_clusters = final_labels
        return final_labels
    
    def get_cluster_statistics(self) -> dict:
        """
        Compute statistics for the final clustering.
        
        Returns:
        --------
        dict
            Cluster-wise statistics
        """
        if self.final_clusters is None:
            raise ValueError("Segmentation not performed. Call segment() first.")
        
        unique_clusters = np.unique(self.final_clusters)
        stats = {
            'n_clusters': len(unique_clusters),
            'cluster_sizes': {
                cluster: np.sum(self.final_clusters == cluster) 
                for cluster in unique_clusters
            },
            'cluster_centroids': {
                cluster: np.mean(self.original_data[self.final_clusters == cluster], axis=0)
                for cluster in unique_clusters
            }
        }
        
        return stats

# Example usage
def example_usage():
    # Generate some sample data
    np.random.seed(42)
    data = np.concatenate([
        np.random.normal(0, 1, (100, 2)),   # Cluster 1
        np.random.normal(5, 1, (100, 2)),   # Cluster 2
        np.random.normal(10, 1, (100, 2))   # Cluster 3
    ])
    
    # Create segmentation instance
    segmenter = IterativeSegmentation(
        data, 
        segmentation_method=KMeans,
        scoring_method=silhouette_score,
        max_clusters=5,
        min_cluster_separation=0.5
    )
    
    # Perform segmentation
    labels = segmenter.segment()
    
    # Get cluster statistics
    stats = segmenter.get_cluster_statistics()
    print("Cluster Statistics:", stats)
```
