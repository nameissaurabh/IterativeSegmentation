```python
class IterativeSegmentation:
    def __init__(
        self, 
        data: np.ndarray, 
        segmentation_method: Callable = KMeans,
        scoring_method: Callable = silhouette_score,
        max_clusters: int = 10,
        min_cluster_separation: float = 0.5
    ):
        """
        Initialize the Iterative Segmentation class.
        
        Parameters:
        -----------
        data : np.ndarray Input data to be segmented
        segmentation_method : Callable, optional Clustering algorithm to use (default is KMeans)
        scoring_method : Callable, optional Metric to evaluate cluster separation (default is silhouette_score)
        max_clusters : int, optional Maximum number of clusters to form (default is 10)
        min_cluster_separation : float, optional Minimum acceptable cluster separation score (default is 0.5)
        """
        # super().__init__()
        self.original_data = data
        self.segmentation_method = segmentation_method
        self.scoring_method = scoring_method
        self.max_clusters = max_clusters
        self.min_cluster_separation = min_cluster_separation
        self.pairwise_score = None
        # Store results
        self.final_clusters = None
        self.cluster_details = []

    def evaluate_clustering(self, labels: np.ndarray, data: np.ndarray) -> float:
        """
        Evaluate the quality of clustering using the specified scoring method.        
        Parameters:
        -----------
        labels : np.ndarray             Cluster labels
        data   : np.ndarray             Data points        
        Returns:
        --------
        float                           Clustering score
        """
        try:
            return self.scoring_method(data, labels)
        except Exception as e:
            print(f"Scoring error: {e}")
            return 0
        
    def parwise_scoring(self, data: np.ndarray, lab_col : str) -> float:
        """
        calculate the pairwise silhoitte score"""
        labels = sorted(data[lab_col].unique())
        pairs = list(itertools.combinations(labels, 2))
        scores = {}
        for i in pairs:
            pair_seg_data = data[data[lab_col].isin(list(i))].reset_index()
            pair_labels = pair_seg_data[lab_col]
            pair_seg_data = pair_seg_data.drop([lab_col], axis = 1)
            scores[i] = self.evaluate_clustering(pair_labels.values, pair_seg_data.values)
            # print(scores)
        
        scores = dict(sorted(scores.items(), key=lambda item: item[1]))
        return scores

    def recursive_segment(self, data: np.ndarray, max_depth: int = 4,
                          threshold = 0.3) -> List[np.ndarray]:
        """
        Recursively segment data, breaking down clusters that are not well-separated.        
        Parameters:
        -----------
        data : np.ndarray                        Data to segment
        current_depth : int, optional            Current recursion depth
        max_depth : int, optional                Maximum recursion depth        
        Returns:
        --------
        List[np.ndarray]                         List of cluster labels
        """
        org_labels = data['labels'].unique()

        # if current_depth >= max_depth or data.shape[0] <= 2:
        #     return [np.zeros(data.shape[0], dtype=int)]
        
        print('Calculating Pairwise score...')
        pairwise_score = self.parwise_scoring(data, 'labels')

        # filter scores for given threshold
        filter_scores = dict(filter(lambda x: x[1] <= threshold, pairwise_score.items()))
        final_score = dict(filter(lambda x: x[1] >= threshold, pairwise_score.items()))
        if filter_scores == {}:
            print('All Clusters are well seperated according to given threshold.. try different threshold or choose base clustring')
            return data
        
        # check the maximum number of clusters
        if len(data['labels'].unique()) >= self.max_clusters:
            print('Clusters are exceeding maximum clusters limit.. making exit')
            return data
        # while not any([data.shape[0] == 1,]):
        
        # do iterative segmentation
        # find segment with low score
        filter_seg = filter_scores.keys()
        # Extract unique elements from the tuple keys
        unique_elements = set()
        for key in filter_seg:
            unique_elements.update(key)

        # Convert the set to a list
        unique_seg_list = list(unique_elements)
    
        # do iterative Seg
        current_depth = 0

        final_data = data.copy()
        
        while True:
            if any([final_data['labels'].nunique() >= self.max_clusters,\
                       filter_scores == {}, current_depth > max_depth]):
                if final_data['labels'].nunique() >= self.max_clusters:
                    print('Exceeding maximum cluster size!!! Breaking...')
                    break
                if  filter_scores == {}:
                    print('Well seperated clusters are found!!! Breaking...')
                    break
                if current_depth > max_depth:
                    print('Exceeding Maximum Depth!!! Breaking...')
                    break
                
            print(f'Doing iteration {current_depth}')
            # create segmentation data
            seg_data = final_data[final_data['labels'].isin(unique_seg_list)].reset_index(drop = True)
            if seg_data.empty:
                print("Segmentation failed due to empty segment data.")
                break
            # remove training data
            final_data = final_data[~final_data['labels'].isin(unique_seg_list)].reset_index(drop = True)
            # seperate labels
            # drop_labels = seg_data['labels'].values
            # drop labels column
            seg_data = seg_data.drop(['labels'], axis = 1)
            
            # get seg algorithm
            seg_algo = self.segmentation_method
            # fit model, Predict labels
            seg_algo.fit(seg_data)
            new_labels = seg_algo.labels_
            # new_labels = [str(i) + itr for i in new_labels] # attched itr for ref
            # attached new labels
            seg_data['labels'] = [str(i) +'_'+ str(current_depth) for i in new_labels]
            # get scores
            print(f'Calculating Pairwise score for {current_depth}...')
            print(seg_data.shape)
            pairwise_score = self.parwise_scoring(seg_data, 'labels')
            # filter scores for given threshold
            filter_scores = dict(filter(lambda x: x[1] <= threshold, pairwise_score.items()))
            # update score
            final_score.update(dict(filter(lambda x: x[1] >= threshold, pairwise_score.items())))
            # find segment with low score
            filter_seg = filter_scores.keys()
            # Extract unique elements from the tuple keys
            unique_elements = set()
            for key in filter_seg:
                unique_elements.update(key)

            # Convert the set to a list
            unique_seg_list = list(unique_elements)
            
            # merge it with the original data
            final_data = pd.concat([final_data, seg_data])
            current_depth += 1
            # merge final score
        self.pairwise_score = final_score
        self.final_clusters = final_data['labels'].to_list()
        return final_data
```
